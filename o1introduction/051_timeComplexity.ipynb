# Time complexity
##$O(n)$

In this sector, we will discuss how to find time complexity directly. And here is a for-loop code below. For this loop, the order of *'for (i=0; i<n; i++)'* and *'statement'* both equal to 1. Therefore, we can only use *'statement'* to identify the time complexity, which is $O(n)$. 
```
for (i=0; i<n; i++)
{
  statement;
}
```
Then, we modify this for-loop a little different from before. The $i$ of this loop is decrementing and the statement will execute for n times as well as $i$ decreases. It is in the order of 1, $O(n)$.
```
for (i=n; i>n; i--)
{
  statement;
}
```
Next, we modify it by increasing 2 for i for each step. The statement will run for $\frac{n}{2}$ times. It is in the order of 1, $O(n)$.
```
for (i=0; i<n; i+2)
{
  statement;
}
```
##$O(n^2)$ 

The time complexity of the firt nested-loop is staightforward, $O(n^2)$.
```
for (i=0; i<n; i++)
{
  for (j=0, j<n; j++)
  {
    statement;
  }
}
```
Now, we replace *'j<n'* with *'j<i'*. Here is the nested-loop code and a table of how many times it excutive for each i. If $i=0$, the nested-loop runs for 1 time, and if $i=2$, and if $i=n$, it runs for n times. So on and so forth. Therefore, the time function is

$f(n) = 1 + 2 + 3 + 4 + ... + n = \frac{n(n+1)}{2}$

, which is in the order of 2, noted as $O(n^2)$.
```
for (i=0; i<n; i++)
{
  for (j=0, j<i; j++)
  {
    statement;
  }
}
```

i | j | # of executive times
--- | --- | ---
0 | stop at 0 | 0
1 | 0, stop at 1 | 1
2 | 0, 1, stop at 2| 2
3 | 0, 1, 2, stop at 3| 3
... | ... | ...
n | ... stop at n| n

##$O(\sqrt{n})$

I post another example to show you the oder of $\frac{1}{2}$. 
For this loop, we do not know how much steps it will take. However, there is a relation between $p$ and $n$, $p < n$. If you work out the value of $p$, you will get $n$. And I draw a table of how $p$ changes and $i$ increase.

Now, we define a new variable k standing for the number of steps the loop will cunsume. And if $i$ go through $k$ steps from 0,  

$ p = \frac{k(k+1)}{2}$ .

From the loop condition, when $p = n$, the loop will stop. So let us make 

$\frac{k(k+1)}{2} = n$,

$k \approx {\sqrt{\frac{n}{2}}}$

The k belongs to the order of $\frac{1}{2}$, $O(\sqrt{n})$.




```
p = 0
for (i=0; p<n; i++)
{
  p = p + i;
}
```

i | p
--- | ---
0 | 0
1 | 0 + 1 = 1
2 | 0 + 1 + 2 = 3
3 | 0 + 1 + 2 + 3 = 6
... | ...
k | 0 + 1 + 2 + 3 + ... + k = $\frac{k(k+1)}{2}$
